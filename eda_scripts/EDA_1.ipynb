# 가설 1. 고객 가구의 자녀 수에 따라 지난 2년간 과자에 지출한 금액이 차이가 있을 것이다

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marketing_data = pd.read_csv('./marketing_campaign.csv', sep='\t')
print(marketing_data.head(5))
print(marketing_data.dtypes)
```

```python
plt.figure(figsize=(20,10))
marketing_data['Children'] = (marketing_data['Kidhome'] + marketing_data['Teenhome']).astype(str)
marketing_data = marketing_data.sort_values(by='Children', ascending=False)
sns.boxplot(data=marketing_data, x='MntWines', y='Children', ax=plt.subplot2grid((2,3), (0,0)))
sns.boxplot(data=marketing_data, x='MntFruits', y='Children', ax=plt.subplot2grid((2,3), (0,1)))
sns.boxplot(data=marketing_data, x='MntMeatProducts', y='Children', ax=plt.subplot2grid((2,3), (0,2)))
sns.boxplot(data=marketing_data, x='MntFishProducts', y='Children', ax=plt.subplot2grid((2,3), (1,0)))
sns.boxplot(data=marketing_data, x='MntSweetProducts', y='Children', ax=plt.subplot2grid((2,3), (1,1)))
sns.boxplot(data=marketing_data, x='MntGoldProds', y='Children', ax=plt.subplot2grid((2,3), (1,2)))
plt.show()
```
